## Funciones y límites

### Funciones

Recordando rápidamente qué es una función:

Es una relación donde a cada elemento de un conjunto $A$ se le asigna un elemento de un conjunto $B$ siempre que se vinculen de manera única y exclusiva. En otras palabras, a cada valor del dominio, le corresponde **uno y sólo un valor del rango**.

* **Dominio:** Conjunto de los elementos que definen la función, es decir, los elementos que se van a asociar con otro conjunto, usualmente denotado con $x$.

* **Rango:** Es el conjunto de elementos que son el resultado de la asociación del dominio bajo la relación, usualmente denotado por $y$.

![](https://i.imgur.com/F24Np1f.png)

En una función, entra un elemento $x$ y sale un elemento $y$:

$$f(x)=y$$

> En el caso donde a cada valor del dominio le corresponde uno o más elementos del rango decimos que tenemos una **relación**. Por ejemplo, al definir una relación para describir un círculo en geometría analítica donde tenemos una relación, no una función.

#### Funciones lineales

Para este ejemplo recordemos las **funciones lineales** (rectas) que se definen de la forma:

$$f(x) = mx+b$$

donde $m$ y $b$ $\in \mathbb{R}$. 

$m$ es la pendiente y $b$ índica en que valor intersecciona la recta con el eje $y$.}

$m$ puede ser calculada por:

$$m = \frac{y_2 - y_1}{x_2 - x_1}$$

Donde $x_n$ y $y_n$ representan las coordenadas de 2 puntos cualesquiera pertenecientes a la recta.

El dominio de nuestra función es:

$$\text{Dom}_f=(-\infty, \infty) $$

Y su imagen o rango:

$$\text{Im}_f = (-\infty, \infty) $$

#### Función logarítmica 

Esta función está relacionada por la **relación**:

$$\log_{b}{x} = n \Longleftrightarrow x = b^n$$

donde:

* $b$ es la base.
* $n$ es el exponente al que está elevada la base.
* $x$ es el resultado de elevar a la base $b$ del exponente $n$.

**Ejemplo:**

Teniendo $b=2$ y $n=8$, entonces:

$$2^8 = 256$$

Por lo que $x=256$. Calculando el logaritmo de base 2 de $x$ tendríamos:

$$\log_{2}{256}=8$$

```python
def f(x):
    return np.log2(x)


def graph(equation):
    fig, ax = plt.subplots()
    ax.plot(x,y, color='#6F10D5')
    plt.xlim(-1, 10)
    plt.ylim(-10, 10)
    plt.legend(['log2(x)'])
    ax.grid()
    ax.axhline(y=0, color='#000000')
    ax.axvline(x=0, color='#000000')  
    plt.title(f'Gráfica logarítmica')
    
if __name__ == '__main__':
    res = 10000
    x = np.linspace(0.0000000000000001, 10.0, num=res)
    y = f(x)
    display_graph = graph(y)
```

#### Funciones a trozos

Son funciones que tienen diferentes valores definidos por un intervalo. Por ejemplo la función escalón de Heaviside:

$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```python
def f(x):
    # Obtenemos un vector lleno de 0's
    y = np.zeros(len(x))
    # Obtenemos el índice y los valores para rellenar con 1's
    for idx, x in enumerate(x):
        if x>= 0:
            y[idx] = 1
    return y


def graph(equation):
    fig, ax = plt.subplots()
    ax.plot(x,y, color='#D30FB8')
    plt.xlim(-3, 3)
    plt.ylim(-1, 3)
    plt.legend(['H(x)'])
    ax.grid()
    ax.axvline(x=0, color='#000000') 
    plt.plot(0, marker = 'o', ms = 15, mfc ='w')
    plt.plot(1, marker = 'o', ms = 15, mfc ='#FF00FF')
    plt.title(f'Gráfica de una función a trozos')
    
if __name__ == '__main__':
    res = 10000
    # Como es una variable usamos un vector 
    # Establecemos el intervalo y cuandos elementos queremos (res)
    x = np.linspace(-10.0, 10.0, num=res)
    y = f(x)
    display_graph = graph(y)
```

### ¿Cómo manipular funciones?

#### Desplazamientos verticales y horizontales

Siendo $c$ una constante mayor a 0, entonces la gráfica:

* $y=f(x) + c$ se desplaza $c$ unidades hacia arriba.
* $y=f(x) - c$ se desplaza $c$ unidades hacia abajo.
* $y=f(x - c) $ se desplaza $c$ unidades hacia la derecha.
* $y=f(x+ c) $ se desplaza $c$ unidades hacia la izquierda.

```python
def fs(x,a):
    # Establecemos la relación algebraica
    return x**2, (x**2)+a, (x**2)-a, (x-a)**2,(x+a)**2


def graph(f1, f2, f3, f4, f5, x):
    fig, ax = plt.subplots()
    plt.xlim(-15, 15)
    plt.ylim(-10, 100)
    ax.plot(x,f1, color='#D43434')
    ax.plot(x,f2, color='#FFB900')
    ax.plot(x,f3, color='#0821F3')
    ax.plot(x,f4, color='#7615DC')
    ax.plot(x,f5, color='#FF03E0')
    plt.legend(['x^2','x^2+a: c unidades hacia arriba ','x^2-a: c unidades hacia abajo.',
                '(x-a)^2: c unidades hacia la derecha', '(x+a)^2: c unidades hacia la izquierda'],
               bbox_to_anchor=(1, 1), loc='upper left')
    ax.grid()
    ax.axhline(y=0, color='#000000')
    ax.axvline(x=0, color='#000000')  
    plt.title("Desplazamientos verticales y horizontales")
    
if __name__ == '__main__':
    res = 10000
    a = int(input('Ingresa el valor de la constante a: '))
    x = np.linspace(-50.0, 50.0, num=res)
    f1, f2, f3, f4, f5 = fs(x,a)
    display_graph = graph(f1, f2, f3, f4, f5, x)
```

#### Alargamientos y compresiones

* $y=c\cdot f(x)$ alarga la gráfica verticalmente por un factor de $c$ unidades.
* $y=\frac{1}{c} \cdot f(x)$ comprime la gráfica verticalmente por un factor de $c$ unidades.
* $y=f(c \cdot x)$ comprime la gráfica horizontalmente por un factor de $c$ unidades.

### Composición de funciones

Cuando hablamos de composición de funciones estamos haciendo referencia al resultado de aplicar 2 o más funciones sucesivas sobre un mismo elemento $x$.

Siendo $f$:

$$\begin{aligned}
f: X \longrightarrow  Y&  \\ \\
x \longmapsto y& = f(x) 
\end{aligned}$$

y $g$:

$$\begin{aligned}
g: Y \longrightarrow  Z&  \\ \\
y \longmapsto z& = g(x) 
\end{aligned}$$

Donde la imagen o **rango** de $f$ **está contenido en el dominio de $g$**. Definimos la composición de $f$ y $g$ como:

$$
\begin{aligned}
(g \circ f): X \longrightarrow  Z&  \\ \\
x \longmapsto z& = g(f(x)) 
\end{aligned}
$$

> Notemos que se nombran en el orden de aplicación de la variable, no en el orden sucesivo de la representación.

Y a todos los elementos de $X$ se le asocia un elemento de $Z$ según: $z=g(f(x))$:

$$\begin{aligned}
X \longrightarrow Y  \longrightarrow  &Z  \\ \\
x \longmapsto y  =f(x) \longmapsto &z = g(f(x)) 
\end{aligned}$$

Se puede representar de manera gráfica usando la categoría de conjuntos, mediante un diagrama conmutativo

![](https://upload.wikimedia.org/wikipedia/commons/thumb/e/ef/Commutative_diagram_for_morphism.svg/200px-Commutative_diagram_for_morphism.svg.png)

O de una manera más intuitiva como la siguiente

![](https://www.universoformulas.com/imagenes/matematicas/analisis/composicion-funciones.jpg)

> **NOTA:** El símbolo $\circ$ se llama **operador de anillo**.

```python
def f(x):
    return x**2

def g(x):
    return np.sin(x)

def graph(f1, f2, x):
    fig, ax = plt.subplots()
    plt.xlim(-4, 4)
    plt.ylim(-1, 3)
    ax.grid()
    ax.plot(x,f1, color='#F93800')
    ax.plot(x,f2, color='#011EFE')
    plt.legend(['f(x)','g(x)'])
    plt.title("Funciones originales")
    ax.axhline(y=0, color='#000000')
    ax.axvline(x=0, color='#000000')
    
def graph2(f1, f2, x):
    fig, ax = plt.subplots()
    plt.xlim(-4, 4)
    plt.ylim(-1, 3)
    ax.grid()
    ax.plot(x,f1, color='#10CB0D')
    ax.plot(x,f2, color='#3E00FF')
    plt.legend(['fog','gof'])
    plt.title("Composición de funciones")
    ax.axhline(y=0, color='#000000')
    ax.axvline(x=0, color='#000000')
    
    
if __name__ == '__main__':
    res = 10000
    x = np.linspace(-50.0, 50.0, num=res)
    y = f(x)
    z = g(x)
    f_o_g = f(g(x))
    g_o_f = g(f(x))
    display_graph = graph(y, z, x)
    display_graph = graph2(f_o_g, g_o_f, x)
```

### Repaso de funciones reales

¿Por qué les llamamos funciones reales?

Se les llaman así por que **tanto su dominio como el codominio (rango)** están contenidos en el conjunto de los números reales $\mathbb{R}$. Es decir, este conjunto que contiene tanto a los racionales como a los irracionales.

Estas funciones tienen una serie de características.

#### Función par

Una función es par si cumple la siguiente relación **a lo largo de su dominio**:

$$f(-x) = f(x)$$

Que de hecho, nos está diciendo que una función es par si es simétrica al eje $y$. Ya que obtenemos los mismos valores de $y$ para diferentes valores de $x$.

Un ejemplo de esto es la parábola:

```python
def f(x,p):
    return 4*p*x**2 


def graph(equation):
    fig, ax = plt.subplots()
    plt.xlim(-2, 2)
    plt.ylim(-1, 10)
    ax.plot(x,y, color='#E52097')
    plt.legend([f'eq'])
    ax.grid()
    ax.axhline(y=0, color='#000000')
    ax.axvline(x=0, color='#000000')  
    plt.title(f'Gráfica de una parábola y=4*{p}*x^2')
    ax.axhline(y=2, color='#680BC3', linestyle = 'dotted')
    ax.axhline(y=4, color='#680BC3', linestyle = 'dotted')
    ax.axhline(y=6, color='#680BC3', linestyle = 'dotted')
    ax.axhline(y=8, color='#680BC3', linestyle = 'dotted')
    
if __name__ == '__main__':
    res = 10000
    p = int(input('Ingresa el valor p de la parábola: '))
    x = np.linspace(-10.0, 10.0, num=res)
    y = f(x,p)
    display_graph = graph(y)
```

#### Funciones monótonas

Estas funciones son útiles de reconocer o analizar debido a que **nos permite saber si una función crece o decrece en alguno de sus intervalos**. Que algo sea monótono se refiere a que no tiene variaciones. Entonces, las funciones monótonas son aquellas dentro de un intervalo $I \in \mathbb{R}$, que **cumplen con alguna** de las siguientes propiedades:

**1. La función es monótona y estrictamente creciente:**

$$\text{si para todo } x_1, x_2 \in I: x_1 < x_2 \Longleftrightarrow f(x_1) < f(x_2)$$

> Si para todo $x_1$ y $x_2$ que pertenecen al intervalo $I$, tal que $x_1$ sea menor a $x_2$, si y solo si $f(x_1)$ es menor que $f(x_2)$.

Lo que nos está diciendo esta definición es que $x_1$ siempre tiene que ser menor que $x_2$ en nuestro intervalo $I$, y que al evaluar $x_2$ en la función obtendremos un número mayor que al evaluar $x_1$ (es creciente).

La siguientes definiciones se leen de una forma muy similar.

**2. La función es estrictamente decreciente:**

$$\text{si para todo } x_1, x_2 \in I: x_1 < x_2 \Longleftrightarrow f(x_1) > f(x_2)$$

**3. La función es monótona y creciente:**

$$\text{si para todo } x_1, x_2 \in I: x_1 < x_2 \Longleftrightarrow f(x_1) \leq f(x_2)$$

**4. La función es monótona y decreciente:**

$$\text{si para todo } x_1, x_2 \in I: x_1 < x_2 \Longleftrightarrow f(x_1) \geq f(x_2)$$

#### Funciones periódicas

$$f(x) = f(x+T) , T \neq 0$$

Por ejemplo, las funciones seno y coseno son funciones periódicas con un periodo $T=2 \pi$. Es decir, después de "recorrer" $2\pi$ a través del dominio obtendremos el mismo valor en el rango.

### ¿Cómo se compone una neurona?

Una neurona artificial o neurona de McCulloch-Pitts , es una unidad de cálculo que intenta modelar el comportamiento de una neurona "natural", similares a las que constituyen al cerebro humano. En su nivel más fundamental no es más que una función matemática, de manera que recibe inputs o **estímulos** para poder realizar su operación o **activarse**.

![](https://magiquo.com/wp-content/uploads/2019/11/neurona.png)

El resultado del cálculo en una neurona es realizar una **suma ponderada de las entradas** (cada entrada tiene distinto peso $w$), seguida de la aplicación de una función no lineal conocida como **función de activación**.

El nuestras entradas tendremos los valores $x_1, x_2, ..., x_n$. Cada una asociada con un peso $w_1, w_2, ..., w_n$. Además de que tendremos un parámetro $b$ que representa nuestro threshold, que es un rango que le da cierta flexibilidad a nuestra neurona para saber cómo ajusta el valor.

La suma ponderada se expresaría como:

$$y = f(x)= w_1x_1+ w_2x_2+ ... w_nx_n + b $$

también podemos encontrar a esta suma ponderada denotada por $red$.

Gráficamente lo que tenemos es una linea recta, donde nuestra combinación lineal $ w_1x_1+ w_2x_2+ ... w_nx_n $ toma el lugar de nuestra parte $mx$ de la ecuación general de la recta, mientras que el término $b$ (bias) representa la intersección con el eje $y$.

![](https://i.imgur.com/bG5Te9n.png)

Matemáticamente la podemos expresar cómo:

$$y = f(x)= \sum_{i=1}^{n}{w_ix_i}  + b $$

> b está fuera de la sumatoria.

¿Qué pasa cuando tenemos más de una neurona conectada?

Entonces tendremos neuronas interconectadas, donde cada una de ellas efectuará una función $f(x)$ y gráficamente obtendremos una gráfica, pero ¿Qué obtenemos después de aplicar varias funciones?

Aquí entra en juego la **composición de funciones**, ya que lo que estamos haciendo no es más que realizar una función, para después realizar una función de la función y así sucesivamente hasta obtener una salida; así que al final, obtendríamos una linea recta (en el caso donde todas las funciones son lineales) sin importar cuántas veces hagamos la composición.

![](https://i.imgur.com/3SGzkda.png)

Podemos mejorar el comportamiento de nuestra red de neuronas utilizando funciones de activación (funciones no lineales).

### Funciones de activación en una neurona $\phi$

Ya tenemos una función dentro de cada neurona que corresponderá con la suma ponderada de sus entradas junto con el bias que nos permite un poco de flexibilidad para la función, pero aún tenemos que encontrar la forma de saber en que punto la neurona está activada o no, estableciendo un rango de valores de umbral.

Recordemos que dentro de cada neurona tendremos:

$$y = f(x)= \sum_{i=1}^{n}{w_ix_i}  + b $$

Nuestras funciones de activación, representadas por la letra $\phi$, son funciones no lineales que nos ayudan a establecer en que momento se han activado nuestras neuronas. 

Usualmente de describen de la forma:

$$s(u)= 
\begin{cases}
1, u \geq 0 \\
0, u < 0
\end{cases}$$

> Se activa si $u \geq 0 $. Si nos damos cuenta, esta es la función de Heaviside que graficamos arriba.

Entonces al conectar nuestras neuronas obtendremos una composición de funciones de la forma:

$$x_n \longrightarrow f(x) \longrightarrow \phi \longrightarrow u$$

Ingresa nuestro valor de entrada $x_n$ en nuestra función $f(x)$, lo cual nos regresa una función lineal (linea recta); después este resultado ingresa en nuestra función de activación $\phi$ que nos regresa una **función no lineal**. 

¿Y por qué es importante que obtengamos como resultado una función no lineal? Bueno, eso nos permite adaptar nuestro algoritmo a comportamientos más complejos partiendo de comportamientos lineales, de forma que vamos construyendo modelos que representen funciones que en un inicio no podríamos plantear para el problema por nuestra propia cuenta. En otras palabras, muchas veces nuestros datos están dispersos y para nosotros nos es muy difícil encontrar funciones que se adapten al comportamiento de nuestros datos, por ello recurrimos a la construcción de redes neuronales que hagan esto por nosotros.

#### Función de Heaviside o función de paso escalonado (escalón binario)

Esta función solo puede tomar 2 valores discretos en su rango $\{0,1\}$, de forma que no hay "pasos intermedios" 

Recordando la ecuación y su gráfica:

$$H(x)= 
\begin{cases}
0, x < 0\\
1, x \geq 0 
\end{cases}$$

Es decir, vamos a recibir una señal de activación si rebasamos el "umbral" con un valor no negativo.

```python
def f(x):
    y = np.zeros(len(x))
    for idx, x in enumerate(x):
        if x>= 0:
            y[idx] = 1
    return y


def graph(equation):
    fig, ax = plt.subplots()
    ax.plot(x,y, color='#E80000')
    plt.plot(0, marker = 'o', ms = 15, mfc ='w')
    plt.plot(1, marker = 'o', ms = 15, mfc ='r')
    plt.xlim(-2, 2)
    plt.ylim(-1, 2)
    plt.legend(['H(x)'])
    ax.grid()
    ax.axvline(x=0, color='#000000')  
    plt.title(f'Función de Heaviside')
    
if __name__ == '__main__':
    res = 10000
    x = np.linspace(-10.0, 10.0, num=res)
    y = f(x)
    display_graph = graph(y)
```

#### Función sigmoide

Muchos proceso naturales y curvas de aprendizaje de sistemas complejos muestran un progresión temporal, con niveles bajos al inicio hasta acercarse a un clímax transcurrido un cierto tiempo; la transición se produce en una región caracterizada por una fuerte aceleración intermedia. La **función sigmoide** nos permite describir esta evolución.

Está función va a tener los intervalos $(0,1)$ para su rango (intervalos abiertos, es decir, no los toca).

Es decir:

$$\begin{aligned}
\lim_{t \rightarrow \infty }{P(t)}= & 1 \\ \\
\lim_{t \rightarrow -\infty }{P(t)}= & 0
\end{aligned}$$

y se describe mediante la fórmula, como un caso particular de la función logística:

$$P(t)= \frac{1}{1+e^{-t}}$$

#### Tangente hiperbólica

Esta es otra función que se suelte utilizar en las funciones de activación. Esta representa un número real $x$ designado mediante $\tanh x$ y se define como el cociente entre el seno hiperbólico y el coseno hiperbólico del número real $x$.

Se escribe como:

$$\tanh x = \frac{\sinh x}{\cosh x}$$

y de acuerdo con las definiciones de seno y coseno hiperbólico, obtenemos:

$$\tanh x = \frac{e^x - e^{-x}}{e^x + e^{-x}}$$

En donde su **diminio** va desde $(\infty, +\infty)$ y su rango desde $(-1,1)$.

```python
def tanh(x):
    return (np.e**x - np.e**(-x))/(np.e**x + np.e**(-x))

def graph(equation):
    fig, ax = plt.subplots()
    ax.plot(t,y1, color='#1848BF')
    plt.xlim(-6, 6)
    plt.ylim(-1.5, 1.5)
    plt.legend(['tanh(x)'])
    ax.grid()
    ax.axvline(x=0, color='#000000')  
    ax.axhline(y=0, color='#000000') 
    plt.title(f'Tangente hiperbólica')
    
if __name__ == '__main__':
    res = 10000
    x = np.linspace(-10.0, 10.0, num=res)
    y1 = tanh(x)
    display_graph = graph(y1)
```

A diferencia de la función sigmoide, en este caso tenemos un comportamiento un poco más rápido para los intervalos $[-2,2]$ como podemos ver en la figura. Además de que esta sí contempla valores negativos para su rango.

#### ReLu (rectified linear unit)

También conocido como **rectificador** es una función de activación definida cómo:

$$f(x)= \text{max}(0,x)=
\begin{cases}
0, x \leq 0 \\ \\
x, x>0
\end{cases}$$

con un rango de $\big[ 0,\infty\big )$.

también es conocida como una función **rampa** y es análoga a la rectificación de media onda en electrónica. Se utiliza mucho en redes convolucionales obteniendo resultados de menor error que los generados con una función logística. 

Las unidades que emplean el rectificador son conocidas como **unidad lineal rectificada**.

Una aproximación suave del rectificador es la función analítica:

$$f(x)= log(1+e^x)$$

Ahora vamos a definir de que formas podemos calcular los errores que nos interesan conocer.

#### Error cuadrático medio - ECM

En estadística, el error cuadrático medio de un estimador **mide el promedio de errores al cuadrado**, es decir, la diferencia entre el estimador y lo que se estima. El ESCM es una función de riesgo, correspondiente al valor esperado de la pérdida del error al cuadrado o pérdida cuadrática. La diferencia se produce debido a la aleatoriedad o porque el estimador no tiene en cuenta la información que podría producir una estimación más precisa.

El ECM incorpora tanto la **varianza** del estimador así como su **sesgo**.

Si $\hat y$ es un vector de $n$ predicciones y $y$ es el vector de los verdaderos valores, entonces el ECM del predictor es:

$$\text{RMSE} = \frac{1}{n}\cdot \sum_{i=1}^{n}{(\hat y_i - y_i)^2} $$

Que al final de cuentas no es más que una regresión lineal. Si quisieramos graficar la linea obtendríamos sus parámetros de ls siguiente forma:

$$m = \frac{n \sum{xy-(\sum{x})(\sum{y})}}{n\sum{x^2}-(\sum{x)^2}}$$

$$b = \frac{\sum{y}- m(\sum{x})}{n} $$

```python
import matplotlib.pyplot as plt
import numpy as np
import math

# Vamos a hacer la linea recta que correspondera a y_hat
# Ya que ahí efectuamos la evaluación del erro predicho
def calc_m_b(data):
    # obtenemos n
    n = len(data)
    # obtenemos la suma de xy
    # Y sum x y sum y
    sum_xy = 0
    sum_x = 0
    sum_y = 0
    sum_x2 = 0
    for i in data:
        sum_xy += (i[0] * i[1])
        sum_x  += (i[0])
        sum_y  += (i[1])
        sum_x2 += (i[0])**2
    # Calculamos la pendiente
    m = (n * sum_xy - (sum_x * sum_y)) / ( (n * sum_x2) - (sum_x)**2 )
    # Ahora calculemos  b
    b = (sum_y - (m * sum_x) ) / (n)
    #Devolvemos valores
    return m,b
    
    
def y_hat_line(x,m,b):   
    return x*m + b

def ecm(data,m,b):
    n= len(data)
    y_hat = []
    add = 0
    for i in data:
        y_hat.append(y_hat_line(i[0],m,b) )
    #print(f'y_hat values:{y_hat}\n')
    
    for i in range(n):
        add += (( y_hat[i]) - (data[i][1]))**2
        print(f'Predicción {i+1}: {y_hat[i]}')
        print(f'Dato real {i+1}: {data[i][1]}')
    return (1/n)*add
    
def graph(x,data,line):
    #Creamos el cuadrito
    fig, ax = plt.subplots()
    ax.grid()
    plt.title('Función de coste')
    plt.xlabel("Gasto en publicidad")
    plt.ylabel("Ventas")
    ax.plot(x,line)
    for i in data:
        plt.plot(i[0], i[1], 'o')
    plt.legend(['Regresión lineal'])
    
def main():
    # Datos
    res = 10000
    x = np.linspace(0.0, 70.0, num=res)
    data = [[0,100],[20,150],[30,200],[40,180],[50,250],[60,230]]
    m,b = calc_m_b(data)
    print(f'\nPendiente: {m}')
    print(f'Intersección: {b}')
    print(f'-------------------------------------')
    line = y_hat_line(x,m,b)
    ECM = ecm(data,m,b)
    graph(x,data, line)
    print(f'-------------------------------------')
    print(f'Error Cuadrático Medio: {ECM}\n')
    
    
if __name__ == '__main__':
    main()
```

### Límites

En en análisis real y complejo, el concepto de límite es esencial para formalizar la noción intuitiva de aproximación hacía un punto concreto de una sucesión o una función a medida que los parámetros de esa sucesión o función se acercan a un determinado valor.

Cuando hablamos de límite **en una sucesión** nos referimos a que los términos de la sucesión se aproximan arbitrariamente a un único número o punto $L$, si es que existe; para valores grandes de $n$. 

Formalmente, se dice que la sucesión $a_n$ **tiende hasta su límite L**, o que **converge** o **es convergente** a $L$, y se denota como:

$$\lim_{n \rightarrow \infty}{a_n} = L$$

si y solo si para todo valor real $\epsilon > 0$ se puede encontrar un número natural $N$ tal que todos los términos de la sucesión, a partir de un cierto valor natural $n$ mayor que $N$, se acerquen a $L$ **cuando $n$ crezca ilimitadamente**. 

Para el caso de **funciones**, podemos definir el límite de una manera muy similar al caso de las sucesiones. Aquí decimos que los valores de la función que puede tomar dentro de un **intervalo** o **radio de convergencia** se van aproximando a un punto fijado $c$ o **punto de acumulación**, independientemente de que este pertenezca al dominio de la función. 

Decimos que el límite de una función $f(x)$ cuando $x$ tiende a $c$ es $L$, y se escribe como:

$$\lim_{x \rightarrow c}{f(x)} = L$$

Para un mayor rigor matemático se utiliza la definición **épslilon-delta** del límite, que es más estricta y convierte al límite en una gran herramienta del análisis real.

> "El límite de $f(x)$ cuando $x$ tiende a $c$ es igual a $L$ si y sólo si para todo número real $\epsilon$ mayor que cero existe un número real $\epsilon$ mayor que cero tal que si la distancia entre $x$ y $c$ es menor que $\epsilon$, entonces la distancia entre la imagen de $x$ y $L$ es menor que $\epsilon$ unidades".

Para entender qué son los límites, consideremos el siguiente ejemplo donde tenemos la función $f(x)=x+2$.


![](https://i.imgur.com/JwqAItQ.png)

El límite de $f$ en $x=3$ es el valor al cual se aproxima $f$ a medida que estamos más y más proximos a $x=3$. Gráficamente interpretamos a este punto como el valor de $y$ al que se acerca nuestra gráfica de $f$ para dicho valor de $x$.

Por ejemplo, imaginemos que partimos del punto $(1,3)$ y nos movemos en la gráfica hasta estar muy muy cerca de $x=3$, pero **sin llegar a tocarlo**, entonces nuestro valor de $y$ tiende a acercarse a 5 cuando nos aproximamos por la derecha.

![](https://cdn.kastatic.org/ka-perseus-images/7f460d0a673f5af283702fccf1d9da7462a48a5a.gif)

De la misma forma, podemos probar aproximandonos por la derecha, por ejemplo, desde el punto $(5,7)$ y nos movemos hasta estar muy cerca de $x=3$. Podemos ver que también nos estamos aproximando a $y=5$.

![](https://cdn.kastatic.org/ka-perseus-images/11ab15823702ab90d5198a2dcd37d65d52627e7b.gif)

Por esta razón decimos que el límite de $f$ en $x=3$ es 5.

La belleza de los límites radica en que no dependen del valor real de la función en ese punto, puede que tengamos el caso donde la función no este definida en un punto $x$, pero si podemos definir el límite cuando nos aproximamos a ese valor.

¿Qué pasa cuando queremos conocer el comportamiento de nuestra función para valores inimaginablemente pequeños o grandes? Para ello introducimos el concepto de **infinito** que no es más que un concepto para hacer referencia a cantidades de una magnitud tan enorme tal que nisiquiera podríamos imaginarla.

$$\lim_{x \rightarrow \infty}{f(x)} = L$$

Este concepto nos ayuda mucho al entender que funciones crecen o disminuyen "más rápido" que otras, ya que evaluamos cuando estas funciones reciben valores muy grandes ($\infty$) o muy pequeños ($-\infty$) y observamos sus comportamientos:

Así que podemos establecer el crecimiento o la escala de estas funciones en función de su infinito:

![](https://i.imgur.com/Z0PdVNn.png)

Tenemos casos donde nuestra gráficas tienden a acercarse a una linea recta, de forma que la distancia entre la gráfica y la línea tiende a 0 a medida que extendemos nuestra gráfica, pero **nunca se tocan entre sí**.

Por ejemplo, tomemos de ejemplo la siguiente gráfica:

![](https://i.imgur.com/shTeh5z.png)

Si quisieramos evaluar la función en el punto $x=0$ estaríamos dividiendo por 0, cosa que no está definida en matemáticas. Por lo tanto, nuestra gráfica tampoco estará definida en ese punto. 

Entonces, no podremos evaluar el punto $x=0$, pero si vamos a poder estudiar el comportamiento de la gráfica en torno a ese punto a través de su límite cuando $x \rightarrow 0$. Y nos daremos cuenta de que al evaluar el comportamiento de la gráfica acercandonos por la izquierda y por la derecha comenzamos a aproximarnos a una linea vertical imaginaría en $x=0$ que no nos permite terminar de converger en dicho valor. A está linea imaginaria le llamaremos asíntota.

Cuando nos acercamos por la izquierda lo representamos como:

$$\lim_{x \rightarrow a^-}{f(x)} = L$$

Y cuando nos acercamos por la derecha lo representamos como:

$$\lim_{x \rightarrow a^+}{f(x)} = L$$

Decimos que sí el límite por izquierda y por derecha son iguales cuando tienden a $a$, entonces el límite $L$ existe.

## Cálculo diferencial

### La derivada y el problema de la recta tangente

El cálculo se desarrolló a la sombra de 4 problemas en los que estaban trabajando los matemáticos europeos en el siglo $\text{XVII}$.

1. El problema de la recta tangente.
2. El problema de la velocidad y la aceleración.
3. El problema de los máximos y mínimos.
4. El problema del área.

Cada uno de ellos involucra la noción de límite y podría servir como introducción al cálculo.

Aunque para ese entonces Pierre de Fermat, Descartes, Huygens e Issac Barrow habían propuesto soluciones parciales para resolver dichos problemas, fue a Issac Newton y a Leibniz a quienes se le atribuyó la primer solución general de dichos problemas. En particular, la solución de Newton venia de su interés por la refracción de la luz y la óptica.

Primero, ¿a que nos referimos con que una recta sea tangente a una curva en un punto? Tomemos el caro de una circunferencia, la recta tangente en un punto $P$ es la recta perpendicular al radio que pasa por $P$.

![](https://i.imgur.com/SnJXhsZ.png)

Sin embargo, en el resto de curvas no es así de sencillo. En esenia, el problema de encontrar la recta tangente en un punto $P$ se reduce al problema de calcular la *pendiente* de una recta en ese punto. 

Podemos aproximar la pendiente de una recta tangente usando una **recta secante**, es decir, que corta a la curva $f(c)$ en 2 puntos. De forma que pase por el punto $P$ y por otro punto cercano de la curva.

![](https://i.imgur.com/2t7vqvs.png)

Si $(c, f(c))$ es el punto de tangencia y $(c+ \Delta x, f(c+\Delta x))$ es el otro punto de la gráfica de $f$, la pendiente de la recta secante que pasa por ambos puntos se encuentra sustituyendo de la siguiente forma.

Primero, recordando nuestra fórmula de la pendiente que se puede demostrar usando Pitágoras:

$$m=\frac{y_2 - y_1}{x_2 - x_1}$$

Ahora, obtenemos la pendiente de nuestra recta secante:

$$\begin{aligned}
m_\text{sec} &= \frac{f(x + \Delta x) - f(c)}{(c + \Delta x) - c} \\ \\
&= \frac{f(x + \Delta x) - f(c)}{\Delta x} 
\end{aligned}$$

El miembro de la derecha de nuestra ecuación es un **cociente de incremento o de diferencias**, mientras que $\Delta x$ es el **cambio** o incremento en $x$ y el numerador $\Delta y$ es el **cambio** o incremento en $y$.

La belleza de este procedimiento radica en que se pueden obtener más aproximaciones y más precisas de la pendiente de la recta tangente a medida que tomamos puntos de la gráfica más próximos al punto de tangencia $P$ (aproximamos una recta secante a una recta tangente a medida que disminuimos la distancia entre los puntos).

![](https://i.imgur.com/8YFbb3u.png)

Entonces vamos a formalizar este último concepto donde queremos hacer casi nula la distancia entre los puntos, es decir, cuando $\Delta x \rightarrow 0$

Si $f$ está definida en un intervalo abierto que contiene a $c$ y además existe el límite:

$$\lim_{\Delta x \rightarrow 0}{\frac{\Delta y}{\Delta x}}= \lim_{x \rightarrow 0}{\frac{f(x + \Delta x) - f(c)}{\Delta x} } = m$$

entonces, la recta que pasa por $(c, f(c))$ y cuenta con una pendiente $m$ **es la recta tangente** a la gráfica de $f$ en el punto $(c, f(c))$

Y justo de aquí nace la definición de la derivada.

La definición de la recta tangente a una curva **no incluye la posibilidad de una recta tangente vertical** ($m=1$). Para estas se utiliza la siguiente definición.

Si $f$ es continua en $c$ y:

$$\lim_{\Delta x \rightarrow 0}{\frac{f(c + \Delta x) - f(c)}{\Delta x}} = \infty$$

o

$$\lim_{\Delta x \rightarrow 0}{\frac{f(c + \Delta x) - f(c)}{\Delta x}} = -\infty$$

entonces la recta $x=c$, que pasa por $(c,f(c))$ es una **recta tangente vertical** a la gráfica de $f$.

#### Ejemplos

##### Pendiente de la gráfica de una función lineal

Encontrar la pendiente de la gráfica de:

$$f(x)= 2x-3$$

en el punto $(2,1)$.

$$m = \lim_{\Delta x \rightarrow 0}{\frac{\Delta y}{\Delta x}}= \lim_{x \rightarrow 0}{\frac{f(x + \Delta x) - f(c)}{\Delta x}}$$

Sustituimos nuestra coordenada conocida $x=2$ del punto $P$ a  evaluar.

$$m =\lim_{\Delta x \rightarrow 0}{\frac{f(2+ \Delta x) - f(2)}{\Delta x}}$$

Ahora sustituyendo en nuestra función $f$ a $(2+\Delta x)$ en lugar de $x$:

$$\begin{aligned}
m &=\lim_{\Delta x \rightarrow 0}{\frac{[2(2+ \Delta x)-3] - [2(2)-3]}{\Delta x}} \\ \\
& = \lim_{\Delta x \rightarrow 0}{\frac{4 + 2\cdot \Delta x-3 - 4 + 3}{\Delta x}}\\ \\
& = \lim_{\Delta x \rightarrow 0}{\frac{2\cdot \Delta x}{\Delta x}}\\ \\
& = 2 \\ 
\end{aligned}$$

![](https://i.imgur.com/KuhpifS.png)

##### Rectas tangentes a la gráfica de una función no lineal

Calcular las pendientes de las rectas tangentes a la gráfica de:

$$f(x)= x^2 +1$$

en los puntos $(0,1)$ y $(-1,2)$.

Sea $(c, f(c))$ un **punto cualquiera** un punto cualquiera de la gráfica de $f$. La pendiente de la recta tangente a él será:

$$m =  \lim_{x \rightarrow 0}{\frac{f(c + \Delta x) - f(c)}{\Delta x}}$$

Reescribiendo $f(x)$ como $(c +  \Delta x)^2 +1 $:

$$\begin{aligned}
m &=  \lim_{x \rightarrow 0}{\frac{[c^2 + 2c \Delta x + (\Delta x)^2 + 1] - [c^2 +1]}{\Delta x}} \\ \\
& = \lim_{\Delta x \rightarrow 0}{\frac{2c  \Delta x + (\Delta x)^2 }{\Delta x}}\\ \\
& = \lim_{\Delta x \rightarrow 0}{\frac{2c  + \Delta x }{1}}\\ \\
& = 2c \\ 
\end{aligned}$$

### Derivada de una función

El límite utilizado para definir la pendiente de una recta tangente también se utiliza para definir una de las 2 operaciones fundamentales del cálculo, la derivación.

La derivada de $f$ en $x$ está dada por:

$$f\prime (x)= \lim_{\Delta x \rightarrow 0}{\frac{f(x+\Delta x) - f(x)}{\Delta x}}$$

Siempre que exista ese límite. Para todos los $x$ para los que exista este límite, $f\prime$ es una función de $x$.

Es importante recordar que la derivada de una función de $x$ también es una función de $x$. Esta "nueva" función proporciona la pendiente de la recta tangente a la gráfica de $f$ en el punto $(x, f(x))$, siempre que la gráfica tenga una recta tangente en dicho punto.

Decimos que una función es **derivable** en $x$ si existe su derivada en $x$, y decimos que es **derivable en un intervalo abierto (a,b)** si es derivable en todos y cada uno de los puntos de ese intervalo.

$f\prime (x)$ se lee como "f prima de x", pero podemos encontrarnos con otras notaciones como:

$$f\prime(x) = \frac{d_y}{d_x}= y\prime = \frac{d}{dx}[f(x)] = D_x[y] $$
