## Funciones y límites

### Funciones

Recordando rápidamente qué es una función:

Es una relación donde a cada elemento de un conjunto $A$ se le asigna un elemento de un conjunto $B$ siempre que se vinculen de manera única y exclusiva. En otras palabras, a cada valor del dominio, le corresponde **uno y sólo un valor del rango**.

* **Dominio:** Conjunto de los elementos que definen la función, es decir, los elementos que se van a asociar con otro conjunto, usualmente denotado con $x$.

* **Rango:** Es el conjunto de elementos que son el resultado de la asociación del dominio bajo la relación, usualmente denotado por $y$.

![](https://i.imgur.com/F24Np1f.png)

En una función, entra un elemento $x$ y sale un elemento $y$:

$$f(x)=y$$

> En el caso donde a cada valor del dominio le corresponde uno o más elementos del rango decimos que tenemos una **relación**. Por ejemplo, al definir una relación para describir un círculo en geometría analítica donde tenemos una relación, no una función.

#### Funciones lineales

Para este ejemplo recordemos las **funciones lineales** (rectas) que se definen de la forma:

$$f(x) = mx+b$$

donde $m$ y $b$ $\in \mathbb{R}$. 

$m$ es la pendiente y $b$ índica en que valor intersecciona la recta con el eje $y$.}

$m$ puede ser calculada por:

$$m = \frac{y_2 - y_1}{x_2 - x_1}$$

Donde $x_n$ y $y_n$ representan las coordenadas de 2 puntos cualesquiera pertenecientes a la recta.

El dominio de nuestra función es:

$$\text{Dom}_f=(-\infty, \infty) $$

Y su imagen o rango:

$$\text{Im}_f = (-\infty, \infty) $$

#### Función logarítmica 

Esta función está relacionada por la **relación**:

$$\log_{b}{x} = n \Longleftrightarrow x = b^n$$

donde:

* $b$ es la base.
* $n$ es el exponente al que está elevada la base.
* $x$ es el resultado de elevar a la base $b$ del exponente $n$.

**Ejemplo:**

Teniendo $b=2$ y $n=8$, entonces:

$$2^8 = 256$$

Por lo que $x=256$. Calculando el logaritmo de base 2 de $x$ tendríamos:

$$\log_{2}{256}=8$$

```python
def f(x):
    # Establecemos la relación algebraica
    return np.log2(x)


def graph(equation):
    # Obtenemos la figura y los ejes
    fig, ax = plt.subplots()
    # Gráficamos en los ejes
    ax.plot(x,y, color='#6F10D5')
    plt.xlim(-1, 10)
    plt.ylim(-10, 10)
    plt.legend(['log2(x)'])
    # Creamos el la rejilla para que se vea bonis
    ax.grid()
    # Y nuestros ejes
    ax.axhline(y=0, color='#000000')
    ax.axvline(x=0, color='#000000')  
    plt.title(f'Gráfica logarítmica')
    
if __name__ == '__main__':
    # Creamos los puntos para la gráfica
    res = 10000
    # Como es una variable usamos un vector 
    # Establecemos el intervalo y cuandos elementos queremos (res)
    # X tiene que ser mayor a 0
    x = np.linspace(0.0000000000000001, 10.0, num=res)
    y = f(x)
    display_graph = graph(y)
```

#### Funciones a trozos

Son funciones que tienen diferentes valores definidos por un intervalo. Por ejemplo la función escalón de Heaviside:

$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```python
def f(x):
    # Obtenemos un vector lleno de 0's
    y = np.zeros(len(x))
    # Obtenemos el índice y los valores para rellenar con 1's
    for idx, x in enumerate(x):
        if x>= 0:
            y[idx] = 1
    return y


def graph(equation):
    # Obtenemos la figura y los ejes
    fig, ax = plt.subplots()
    # Gráficamos en los ejes
    ax.plot(x,y, color='#D30FB8')
    # Limitamos la porción del plano
    plt.xlim(-5, 5)
    plt.ylim(-1, 5)
    plt.legend(['H(x)'])
    # Creamos el la rejilla para que se vea bonis
    ax.grid()
    # Y nuestros ejes
    ax.axhline(y=0, color='#000000')
    ax.axvline(x=0, color='#000000')  
    plt.title(f'Gráfica de una función a trozos')
    
if __name__ == '__main__':
    # Creamos los puntos para la gráfica
    res = 10000
    # Como es una variable usamos un vector 
    # Establecemos el intervalo y cuandos elementos queremos (res)
    x = np.linspace(-10.0, 10.0, num=res)
    y = f(x)
    display_graph = graph(y)
```

### ¿Cómo manipular funciones?

#### Desplazamientos verticales y horizontales

Siendo $c$ una constante mayor a 0, entonces la gráfica:

* $y=f(x) + c$ se desplaza $c$ unidades hacia arriba.
* $y=f(x) - c$ se desplaza $c$ unidades hacia abajo.
* $y=f(x - c) $ se desplaza $c$ unidades hacia la derecha.
* $y=f(x+ c) $ se desplaza $c$ unidades hacia la izquierda.

```python
def fs(x,a):
    # Establecemos la relación algebraica
    return x**2, (x**2)+a, (x**2)-a, (x-a)**2,(x+a)**2


def graph(f1, f2, f3, f4, f5, x):
    # Obtenemos la figura y los ejes
    fig, ax = plt.subplots()
    plt.xlim(-15, 15)
    plt.ylim(-10, 100)
    # Gráficamos en los ejes
    ax.plot(x,f1, color='#D43434')
    ax.plot(x,f2, color='#FFB900')
    ax.plot(x,f3, color='#0821F3')
    ax.plot(x,f4, color='#7615DC')
    ax.plot(x,f5, color='#FF03E0')
    plt.legend(['x^2','x^2+a: c unidades hacia arriba ','x^2-a: c unidades hacia abajo.',
                '(x-a)^2: c unidades hacia la derecha', '(x+a)^2: c unidades hacia la izquierda'],
               bbox_to_anchor=(1, 1), loc='upper left')
    # Creamos el la rejilla para que se vea bonis
    ax.grid()
    # Y nuestros ejes
    ax.axhline(y=0, color='#000000')
    ax.axvline(x=0, color='#000000')  
    plt.title("Desplazamientos verticales y horizontales")
    
if __name__ == '__main__':
    # Creamos los puntos para la gráfica
    res = 10000
    # Como es una variable usamos un vector 
    # Establecemos el intervalo y cuandos elementos queremos (res)
    a = int(input('Ingresa el valor de la constante a: '))
    x = np.linspace(-50.0, 50.0, num=res)
    f1, f2, f3, f4, f5 = fs(x,a)
    display_graph = graph(f1, f2, f3, f4, f5, x)
```

#### Alargamientos y compresiones

* $y=c\cdot f(x)$ alarga la gráfica verticalmente por un factor de $c$ unidades.
* $y=\frac{1}{c} \cdot f(x)$ comprime la gráfica verticalmente por un factor de $c$ unidades.
* $y=f(c \cdot x)$ comprime la gráfica horizontalmente por un factor de $c$ unidades.
* $y= f(\frac{1}{c}\cdot x )$ alarga la gráfica horizontalmente por un factor de $c$ unidades.

```python
def fs(x,a):
    return np.sin(x), np.sin(x)*a, np.sin(x)*(1/a), np.sin(x*a)**2,np.sin(x*(1/a))**2

def fs2(x,a):
    return x**2, (x**2)*a, (x**2)*(1/a), (x*a)**2,(x*(1/a))**2

def graph(f1, f2, f3, f4, f5, x):
    # Obtenemos la figura y los ejes
    fig, ax = plt.subplots()
    plt.xlim(-3, 3)
    plt.ylim(-3, 3)
    # Gráficamos en los ejes
    ax.plot(x,f1, color='#F90CB8')
    ax.plot(x,f2, color='#197DAD')
    ax.plot(x,f3, color='#2EAD19')
    ax.plot(x,f4, color='#7615DC')
    ax.plot(x,f5, color='#D10909')
    
    plt.legend(['sin(x)','sin(x)*a: se alarga verticalmente',
                'sin(x)*(1/a): se comprime verticalmente',
                'sin(x*a): se comprime horizontalmente',
                'sin(x*(1/a)): se alarga horizontalmente'],
               bbox_to_anchor=(1, 1), loc='upper left')
    
    ax.grid() 
    plt.title("Alargamientos y compresiones a sin(x)")
    
def graph2(f6, f7, f8, f9, f10, x):
    # Obtenemos la figura y los ejes
    fig, ax = plt.subplots()
    plt.xlim(-3, 3)
    plt.ylim(-3, 3)
    # Gráficamos en los ejes
    ax.plot(x,f6, color='#D43434')
    ax.plot(x,f7, color='#FFB900')
    ax.plot(x,f8, color='#0821F3')
    ax.plot(x,f9, color='#7615DC')
    ax.plot(x,f10, color='#FF03E0')
    plt.legend(['x^2','(x^2)*a: se alarga verticalmente',
                '(x^2)*(1/a): se comprime verticalmente',
                '(x*a)^2: se comprime horizontalmente',
                '(x*(1/a))^2: se alarga horizontalmente'],
               bbox_to_anchor=(1, 1), loc='upper left')
    ax.grid()
    ax.axhline(y=0, color='#000000')
    ax.axvline(x=0, color='#000000')  
    plt.title("Alargamientos y compresiones a x^2")
    
    
if __name__ == '__main__':
    # Creamos los puntos para la gráfica
    res = 10000
    # Como es una variable usamos un vector 
    # Establecemos el intervalo y cuandos elementos queremos (res)
    a = int(input('Ingresa el valor de la constante a1: '))
    a2 = int(input('Ingresa el valor de la constante a2: '))
    x = np.linspace(-50.0, 50.0, num=res)
    
    f1, f2, f3, f4, f5 = fs(x,a)
    display_graph = graph(f1, f2, f3, f4, f5, x)
    
    f6, f7, f8, f9, f10 = fs2(x,a2)
    display_graph2 = graph2(f6, f7, f8, f9, f10, x)
```

#### Reflexiones

* $y = -f(x)$ refleja la gráfica respecto al eje x.
* $y = f(-x)$ refleja la gráfica respecto al eje y.

```python
def f(x):
    # Establecemos la relación algebraica
    return -(1/8)*x * 4**(2-x)


def graph(f1, f2, f3, x):
    # Obtenemos la figura y los ejes
    fig, ax = plt.subplots()
    plt.xlim(-3, 3)
    plt.ylim(-3, 3)
    # Gráficamos en los ejes
    ax.plot(x,f1, color='#29D3B7')
    ax.plot(x,f2, color='#FFB900')
    ax.plot(x,f3, color='#C929D3')
    plt.legend(['f(x)','-f(x)*a: se refleja respecto al eje x',
                'f(-x): se refleja respecto al eje y'])
    # Creamos el la rejilla para que se vea bonis
    ax.grid()
    # Y nuestros ejes
    ax.axhline(y=0, color='#000000')
    ax.axvline(x=0, color='#000000')  
    plt.title("Reflexiones")
    
if __name__ == '__main__':
    # Creamos los puntos para la gráfica
    res = 10000
    x = np.linspace(-50.0, 50.0, num=res)
    f1 = f(x)
    f2 = -f(x)
    f3 = f(-x)
    display_graph = graph(f1, f2, f3, x)
```
